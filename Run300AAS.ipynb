```python
import truncate_newboxsizeAAS as trunc
import numpy as np
import os
run_ind = 0
mass_ind = 0
```

```python
%%time
mH = [300, 250, 350]
mixangle = np.array([.00005, .00005/2, .00005/4, .00005/8, .00005/10, .00005/12, .00005/16, .00005/20, .00005/30])
mixangle *= (300/mH[mass_ind])**2

print("Mass = {} MeV; Lifetime = {:.3} s".format(mH[mass_ind],trunc.tH(mH[mass_ind],mixangle[run_ind])*6.58e-22))
    
fn = "{}-{:.4}-Full/".format(mH[mass_ind],mixangle[run_ind])

if not os.path.exists(fn):
    os.makedirs(fn)
else:
    print("Folder {} already exists.  Overwriting any data".format(fn))
trunc.babysteps(mH[mass_ind],mixangle[run_ind],200,100,100,fn+"full")
```
